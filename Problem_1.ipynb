```python
import pandas as pd
from nsetools import Nse     #library to import NSE data
from pprint import pprint
nse = Nse()
new=[]
stocks =('reliance','hdfcbank','adaniports','itc','sbin','ioc','rblbank')
keys = ('symbol','open','dayHigh','dayLow','closePrice','totalTradedValue','deliveryQuantity','deliveryToTradedQuantity')
column_name  = ['Company','Open','High','Low','Close','Volume','Delivery Data','Delivery Percentage']

def stock_data(stock):  #funtion to return Stock Data
    new_data=[]
    data = nse.get_quote(stock)
    for x in keys:
        new_data.append(data[x])
    return(new_data)

for i in stocks:
    a = stock_data(i)
    new.append(a)
df=pd.DataFrame(new,columns=column_name)  #append dataframe with stocks data
df
```
